```python
# For director, cast, and listed_in
# Because there is more than 1 people and categories
# We don't want if people share the same first or last name consider the same person
# or the word that appear in many categories (TV, etc) consider the same category
def separate(texts):
    t = []
    for text in texts.split(','):
        t.append(text.replace(' ', '').lower())
    return ' '.join(t)

def remove_space(texts):
    return texts.replace(' ', '').lower()

def remove_punc(texts):
    return texts.translate(str.maketrans('','',string.punctuation)).lower()
```

```python
import string
new_df['type'] = new_df['type'].apply(remove_space)
new_df['director'] = new_df['director'].apply(separate)
new_df['cast'] = new_df['cast'].apply(separate)
new_df['rating'] = new_df['rating'].apply(remove_space)
new_df['listed_in'] = new_df['listed_in'].apply(separate)
new_df['description'] = new_df['description'].apply(remove_punc)

new_df.head()
```

### Cosine Similarity with TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfid = TfidfVectorizer()
tfid_matrix = tfid.fit_transform(new_df['bag_of_words'])
cosine_sim = cosine_similarity(tfid_matrix, tfid_matrix)
cosine_sim
```

```python
# Later on we will combine with similarity as a column
final_df = df[['title', 'type']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

def recommendation_with_classification_report(title, total_result=5, threshold=0.5):
    # Get the index
    idx = final_df[final_df['title'] == title].index[0]
    
    # Create a new column for similarity, the value is different for each title you input
    final_df['similarity'] = cosine_sim[idx]
    sort_final_df = final_df.sort_values(by='similarity', ascending=False)[1:total_result+1]
    
    # Set a threshold for positive recommendations
    positive_recommendations = sort_final_df['similarity'] > threshold
    
    # Create a binary classification label indicating relevance
    # Assuming relevant items are those with similarity above a certain threshold
    final_df['label'] = (final_df['similarity'] > threshold).astype(int)

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(
        final_df[['similarity']],  # Features
        final_df['label'],          # Target variable
        test_size=0.2,              # 80% training, 20% testing
        random_state=42             # Set random seed for reproducibility
    )
    
    # Select relevant columns for classification report
    y_true = y_test  # Use testing set for evaluation
    y_pred = positive_recommendations.loc[sort_final_df.index].reindex(y_true.index, fill_value=0).astype(int)
    
    # Generate the classification report
    report = classification_report(y_true, y_pred)
    
    print("Classification Report:")
    print(report)
    
    # Print the recommended items
    print("\nRecommended Items:")
    for i, item in enumerate(sort_final_df['title']):
        print('{}. {}'.format(i+1, item))

# Example usage
recommendation_with_classification_report('Breaking Bad')
```

```python
#import the necessary packages
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

#select the relevant columns
new_df = df[['title', 'type', 'director', 'cast', 'rating', 'listed_in', 'description']]
new_df.set_index('title', inplace=True)
new_df['bag_of_words'] = ''

#concatenate the attributes into one string
new_df['bag_of_words'] = new_df.apply(lambda row: ' '.join(str(x) for x in row), axis=1)

#remove punctuation
new_df['bag_of_words'] = new_df['bag_of_words'].apply(remove_punc)

#drop the other columns
new_df.drop(new_df.columns[:-1], axis=1, inplace=True)

#create the TF-IDF matrix
tfid = TfidfVectorizer()
tfid_matrix = tfid.fit_transform(new_df['bag_of_words'])

#feature scaling using fit_transform()
#pass with_mean=False to avoid centering sparse matrix
scaler = StandardScaler(with_mean=False)
X_scaled = scaler.fit_transform(tfid_matrix)

#split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, df['type'], test_size=0.2, random_state=42)

#choose the best kernel for SVM
#you can try different kernels or tune the parameters to improve the results
kernel = 'sigmoid'

#train the SVM model
svm_model = SVC(kernel=kernel)
svm_model.fit(X_train, y_train)

#evaluate the SVM model
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
print(f"Accuracy with SVM and {kernel} kernel: {accuracy}")
print(f"Classification Report with SVM and {kernel} kernel:\n{report}")

#define a function to generate recommendations for a given title
def recommendation_with_svm(title, total_result=5):
    """This function takes a title and a number of results as input, and returns a list of recommended items based on the predicted label and the TF-IDF scores."""
    #transform the title into a TF-IDF vector
    tfid_vector = tfid.transform([title])
    #predict the label for the title using the SVM model
    predicted_label = svm_model.predict(tfid_vector)[0]
    #filter the items based on the predicted label
    filtered_items = df[df['type'] == predicted_label]
    
    try:
        #calculate the TF-IDF scores for the filtered items
        tfid_scores = tfid_matrix[filtered_items.index.astype(int)].dot(tfid_vector.T).toarray().flatten()
        #sort the items by the TF-IDF scores in descending order
        sort_index = tfid_scores.argsort()[::-1]
        sort_scores = tfid_scores[sort_index]
        sort_title = filtered_items['title'].iloc[sort_index]
        
        #create a dataframe with the title and the score of the recommended items
        result = pd.DataFrame({'Title': sort_title, 'Score': sort_scores})
        #select the top k items as the recommendations
        result = result.iloc[1:total_result + 1]
        
        #print the recommendations
        print(f"Recommended Items for {title}:")
        for i, item in enumerate(result['Title']):
            print(f"{i + 1}. {item}")
            
    except KeyError as e:
        #handle the error if no matching items are found
        print(f"Error: {e}. No matching items found for {title}.")
```

# Hybrid Model - SVM with Cosine Similarity & TFIDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

# Function to remove punctuation
def remove_punc(text):
    return ''.join([char for char in text if char.isalnum() or char.isspace()])

# Select relevant columns
new_df = df[['title', 'type', 'director', 'cast', 'rating', 'listed_in', 'description']]
new_df.set_index('title', inplace=True)

# Keep a copy of the 'type' column
new_df['item_type'] = new_df['type']

# Concatenate the attributes into one string
new_df['bag_of_words'] = new_df.apply(lambda row: ' '.join(str(x) for x in row), axis=1)

# Remove punctuation
new_df['bag_of_words'] = new_df['bag_of_words'].apply(remove_punc)

# Drop the other columns
new_df.drop(new_df.columns[:-2], axis=1, inplace=True)

# Create the TF-IDF matrix
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(new_df['bag_of_words'])

# Feature scaling using fit_transform()
scaler = StandardScaler(with_mean=False)
X_scaled = scaler.fit_transform(tfidf_matrix)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, new_df['item_type'], test_size=0.2, random_state=42)

# Choose the best kernel for SVM
kernel = 'sigmoid'

# Train the SVM model
svm_model = SVC(kernel=kernel)
svm_model.fit(X_train, y_train)

# Evaluate the SVM model
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
print(f"Accuracy with SVM and {kernel} kernel: {accuracy}")
print(f"Classification Report with SVM and {kernel} kernel:\n{report}")

# Function to generate recommendations using hybrid model
def hybrid_recommendation(title, new_df, total_result=5):
    # Get the TF-IDF representation for the input title
    title_tfidf = tfidf.transform([remove_punc(' '.join(str(x) for x in new_df.loc[title]))])
    
    # Predict the type using SVM
    predicted_type = svm_model.predict(title_tfidf)[0]
    
    # Filter dataframe by predicted type
    filtered_df = new_df[new_df['item_type'] == predicted_type]
    
    # Convert indices to numeric format
    numeric_indices = [i for i, idx in enumerate(new_df.index) if idx in filtered_df.index]
    
    # Calculate cosine similarity
    similarity_scores = cosine_similarity(title_tfidf, X_scaled[numeric_indices])
    
    # Get indices of top recommendations
    top_indices = np.argsort(similarity_scores[0])[-total_result:][::-1]
    
    # Display recommendations
    print(f"Recommended Items for {title}:")
    for i, idx in enumerate(top_indices):
        print(f"{i + 1}. {new_df.index[numeric_indices[idx]]}")

# Example usage
hybrid_recommendation('Breaking Bad', new_df)
```
